```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
from libsvm3.python.svmutil import *
from libsvm3.python.svm import *
from sklearn.model_selection import KFold
from sklearn.datasets import dump_svmlight_file
from sklearn.datasets import load_svmlight_file

warnings.filterwarnings("ignore")

cov = [[1, 0], [0, 1]]
mean1 = [1, -1]
mean2 = [5, -4]
mean3 = [1, 4]
mean4 = [6, 4.5]
mean5 = [7.5, 0.0]

x1 = np.random.multivariate_normal(mean1, cov, 200).T
x2 = np.random.multivariate_normal(mean2, cov, 200).T
x3 = np.random.multivariate_normal(mean3, cov, 200).T
x4 = np.random.multivariate_normal(mean4, cov, 200).T
x5 = np.random.multivariate_normal(mean5, cov, 200).T

# plt.scatter(x1[0],x1[1],c='r',s=5)
# plt.scatter(x2[0],x2[1],c='b',s=5)
# plt.scatter(x3[0],x3[1],c='k',s=5)
# plt.scatter(x4[0],x4[1],c='g',s=5)
# plt.scatter(x5[0],x5[1],c='m',s=5)
# plt.show()


Train = pd.DataFrame()
Test = pd.DataFrame()

for p, i in enumerate([x1, x2, x3, x4,x5]):
    modo = pd.DataFrame(i).T
    modo = pd.concat([modo, pd.DataFrame(np.ones(200) * p, columns=['Label'])], axis=1)

    train = modo.sample(frac=0.6, axis=0)
    test_index = [n for n in modo.index if n not in train.index]
    test = modo.iloc[test_index]

    Train = Train.append(train)
    Test = Test.append(test)


def change_to_libsvm(Train, name):
    y = Train.Label  # y为数据的label值
    dummy = pd.get_dummies(Train.iloc[:, :2])
    mat = dummy.as_matrix()
    dump_svmlight_file(mat, y, 'svm_output' + name + '.libsvm', zero_based=False)  ##默认为zero_based=True，转换后的字段编号从0开始

    return load_svmlight_file('svm_output' + name + '.libsvm')


Train_x, Train_y = change_to_libsvm(Train, 'train')
Test_x, Test_y = change_to_libsvm(Test, 'test')


#########################################################################################################
#                        任务一  画出正确率随正则化参数的变化曲线

# 生成浮点数列表 [0.1,0.2 .....10]
def floatrange(start, stop, steps):
    return [start + float(i) * (stop - start) / (float(steps) - 1) for i in range(steps)]


goal_1 = pd.DataFrame()
'''
    goal_1 : 含有两个特征，其中

           特征 C  ：表示 正则化参数
           特征 Acc：表示 正确率

    作用 ：存放模型训练后的数据，以备后续使用

'''

# 该循环用于 输入不同的 正则化值 给模型训练 从而得到对应的正确率
for c in floatrange(0.1, 10, 100):
    prob = svm_problem(Train_y, Train_x)
    param = svm_parameter('-t 0  -b 1  -c ' + str(c + 1))
    model = svm_train(prob, param)
    p_label, p_acc, p_val = svm_predict(Test_y, Test_x, model)
    goal_1 = goal_1.append(pd.DataFrame({"C": [c], "Acc": [p_acc[0]]}))

'''
     p_label ：样本的预测类标号
     p_acc   ：分类的正确率、回归的均方根误差、回归的平方相关系数
     p_val   ：一个矩阵包含决策的值或者概率估计

'''


# 导出 正确率随正则化参数的变化 的数据
# goal_1.to_csv("goal_1.csv")

#  画出 正确率的变化图
plt.plot(goal_1["Acc"].values)
plt.show()


##########################################################################################################
#                       任务二  交叉验证选择正则化参数

Train = Train.append(Test)
Best_C = [0, 0]

for c in floatrange(0.1, 10, 100):
    goal_2 = [c, 0]
    kfold = KFold(n_splits=5, shuffle=True, random_state=0)
    for train_index, test_index in kfold.split(Train):
        train, test = Train.iloc[train_index], Train.iloc[test_index]

        Train_x, Train_y = change_to_libsvm(train, 'cv_train')
        Test_x, Test_y = change_to_libsvm(test, 'cv_test')

        prob = svm_problem(Train_y, Train_x)
        param = svm_parameter('-t 0  -b 1  -c ' + str(c + 1))
        model = svm_train(prob, param)
        p_label, p_acc, p_val = svm_predict(Test_y, Test_x, model)

        goal_2[1] += p_acc[0] / 5

    Best_C = goal_2 if goal_2[1] > Best_C[1] else Best_C

print(" 最优正则化参数为 ：" + str(Best_C[0]) + " 正确率为 ：" + str(Best_C[1]))
```
